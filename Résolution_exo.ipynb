```python
import matplotlib.pyplot as plt 
import numpy
import math 
```

```python
def f(x_values):
    A = numpy.exp(-x_values/10)*numpy.sin(x_values)
    return A 
```

# 1) Représentation de la courbe cf [0,10] avec 10 000 points 

```python
start, end = 0, 10

number_of_points = 10000

x_values = numpy.linspace(start, end, number_of_points)
y_values = f(x_values )
```

# 2) Représentation de la coube cf' [0,10]

```python
df_dx = numpy.gradient(y_values, x_values)
```

```python
    plt.figure(figsize=(15,6))
    plt.plot(x_values, y_values, "b")
    plt.plot(x_values, df_dx, "r")
    plt.show()
```

# 3) La moyenne et l'écart type 

```python
x_values = numpy.linspace(3, 7, 100)
y_values = f(x_values )

mean = numpy.mean(y_values)
std  = numpy.std(y_values)

print("la moyenne est de : ", mean)
print("et l'écart type :", std )
```
